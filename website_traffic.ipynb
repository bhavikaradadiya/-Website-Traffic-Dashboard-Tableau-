```python
import pandas as pd

# Load dataset
df = pd.read_csv("/content/website_traffic.csv")

# Rename columns for Tableau-friendliness (no spaces)
df.columns = df.columns.str.strip().str.replace(" ", "_")

# Check for null values
missing_values = df.isnull().sum()

# Ensure correct datatypes
df = df.astype({
    "Page_Views": "int",
    "Session_Duration": "float",
    "Bounce_Rate": "float",
    "Traffic_Source": "category",
    "Time_on_Page": "float",
    "Previous_Visits": "int",
    "Conversion_Rate": "float"
})

# Clip bounce/conversion rates to max 1.0 just in case
df["Bounce_Rate"] = df["Bounce_Rate"].clip(0, 1)
df["Conversion_Rate"] = df["Conversion_Rate"].clip(0, 1)

# Preview cleaned data
df.to_csv("cleaned_website_traffic.csv", index=False)
print("null summary after data cleaning", missing_values)
print("dataset shape:", df.shape)
df.head()
```
